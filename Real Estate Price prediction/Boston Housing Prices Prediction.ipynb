```python
# outlier treatment

def wisker(col):
    q1,q3 = np.percentile(col,[25,75])
    iqr = q3-q1
    lw = q1-1.5*iqr
    uw = q1+1.5*iqr
    return lw,uw
```

```python
for j in ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']:
    lw,uw = wisker(boston[j])
    boston[j] = np.where(boston[j]<lw,lw,boston[j])
    boston[j] = np.where(boston[j]>uw,uw,boston[j])
```

```python
# checking for outliers
for j in ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT', 'MEDV']:
    sns.boxplot(data=boston,x=j)
    plt.show()
```
