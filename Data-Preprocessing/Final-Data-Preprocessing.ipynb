```python
# loading LOAD_DIABETES datasets
df2=pd.DataFrame(load_diabetes().data)
df2
```

```python
# checking for missing value
df2.isna().any()
```

```python
# checking for outlier
sns.boxplot(x=df2.values)
```

```python
# Treating missing values and outliers
# Treating missing values
```

```python
mean_imputer = SimpleImputer(missing_values=np.nan,strategy="mean")
mean_imputer = mean_imputer.fit(df2)
imputed_data = mean_imputer.transform(df2)
df2 = pd.DataFrame(data=imputed_data)
df2
```

```python
# Treating outliers
```

```python
filter = df2.values>0.15
df2_outlier = df2[filter]
df2_outlier
```

```python
sns.boxplot(x=df2_outlier.values)
```
