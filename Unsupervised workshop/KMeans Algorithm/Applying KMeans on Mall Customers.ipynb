```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\Phavour\Downloads\Personal Python-ML project\Simplilearn\Unsupervised workshop\KMeans Algorithm\Mall_Customers.csv")
data
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3,init='k-means++',random_state=42)

y_kmeans = model.fit_predict(data.iloc[:, [2,3]])
```

```python
y_kmeans
```

```python
X = data.iloc[:, [2,3]].values
```

```python
plt.scatter(X[y_kmeans ==0,0], X[y_kmeans ==0,1], s=100, c='red',label='cluster 1')
plt.scatter(X[y_kmeans ==1,0], X[y_kmeans ==1,1], s=100, c='green', label='cluster 2')
plt.scatter(X[y_kmeans ==2,0], X[y_kmeans ==2,1], s=100, c='blue',label='cluster 3')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], s=300, c='yellow',label='centroid')

plt.title('clusters of three')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

```python
wcss = []

for i in range(1,11):
    model = KMeans(n_clusters=i,init='k-means++',random_state=42)
    model.fit(X)
    wcss.append(model.inertia_)

plt.plot(range(1,11),wcss)
plt.title('Elkow methods')
plt.xlabel('members of cluster')
plt.ylabel('wcss')
plt.show()
```

```python
model2 = KMeans(n_clusters=5,init='k-means++',random_state=42)
y_kmeans2 = model2.fit_predict(data.iloc[:, [2,3]])

plt.scatter(X[y_kmeans2 ==0,0], X[y_kmeans2 ==0,1], s=100, c='red',label='cluster 1')
plt.scatter(X[y_kmeans2 ==1,0], X[y_kmeans2 ==1,1], s=100, c='green', label='cluster 2')
plt.scatter(X[y_kmeans2 ==2,0], X[y_kmeans2 ==2,1], s=100, c='blue',label='cluster 3')
plt.scatter(X[y_kmeans2 ==3,0], X[y_kmeans2 ==3,1], s=100, c='cyan', label='cluster 4')
plt.scatter(X[y_kmeans2 ==4,0], X[y_kmeans2 ==4,1], s=100, c='magenta', label='cluster 5')
plt.scatter(model2.cluster_centers_[:, 0], model2.cluster_centers_[:, 1], s=300, c='yellow',label='centroid')

plt.title('clusters of five(5)')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
