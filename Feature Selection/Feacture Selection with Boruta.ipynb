```python
import pandas as pd
import numpy as np
```

```python
URL = pd.read_csv('https://raw.githubusercontent.com/Aditya1001001/English-Premier-League/master/EDA_data.csv')
URL
```

```python
URL.info()
```

```python
URL = URL.drop(['mod name','Season', 'nationality'],axis=1)
```

```python
URL.info()
```

```python
X = URL.drop('Position',axis=1)
y = URL.Position
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=1000,max_depth=5,random_state=1)
classifier.fit(xtrain, ytrain)
```

```python
from sklearn.metrics import accuracy_score
```

```python
classifier.predict(xtest)
```

```python
accuracy_score(ytest, classifier.predict(xtest))
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_transform = le.fit_transform(ytrain)

y_transform
```

```python
from boruta import BorutaPy
```

```python
classifier_B = RandomForestClassifier(n_estimators=1000,max_depth=5,random_state=1)
boruta_sel = BorutaPy(estimator=classifier_B, n_estimators='auto', verbose=2, random_state=1)
```

```python
boruta_sel.fit(np.array(xtrain), np.array(y_transform))
```
