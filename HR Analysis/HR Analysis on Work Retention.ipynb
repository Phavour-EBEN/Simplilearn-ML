```python
retention = data[['satisfaction_level','average_montly_hours','Work_accident','left','promotion_last_5years','salary']]
retention
```

```python
dummies = pd.get_dummies(retention.salary,prefix='salary')
dummies = dummies.drop('salary_medium',axis=1)
dummies
```

```python
retention = retention.drop('salary',axis=1)
retention_df = pd.concat([retention,dummies],axis=1)
retention_df
```

```python
X = retention_df.drop('left',axis=1)
y = retention_df.left
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression()
log_model.fit(x_train, y_train)
```

```python
log_model.score(x_test,y_test)
```

```python
ypred = log_model.predict(x_test)
```
