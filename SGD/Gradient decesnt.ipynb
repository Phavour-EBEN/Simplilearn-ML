```python
import pandas as pd
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```

```python
def Gradient(x, y):
    n_interation = 1000
    learning_rate = 0.008
    n = len(x)
    m_curr = b_curr = 0
    
    for i in range(n_interation):
        y_pred = m_curr*x + b_curr
        md = -(n/2)*sum(x*(y-y_pred))
        bd = -(n/2)*sum((y-y_pred))
        cost = (1/n)*sum([val**2 for val in y-y_pred])
        
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        print(f"m {m_curr}, b {b_curr},cost {cost}, n_iteration {i}")
```

```python
Gradient(x, y)
```
