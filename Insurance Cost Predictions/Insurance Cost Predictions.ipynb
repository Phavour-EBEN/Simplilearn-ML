```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data.info()
```

```python
d1 = data
data.isna().sum()
```

```python
col_to_convert = ['sex','smoker','region']


dummies = pd.get_dummies(d1[col_to_convert],dtype='int64')

```

```python
dummies
```

```python
dummies = dummies.drop(['sex_female','smoker_no','region_southwest'],axis=1)
```

```python
data = data.drop(col_to_convert,axis=1)
```

```python
data = pd.concat([data,dummies],axis=1)
data
```

```python
X = data.drop('charges',axis=1)
y = data['charges']

```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_trian, y_test = train_test_split(X, y, test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(n_jobs=-1)
model.fit(x_train, y_trian)
```

```python
from sklearn.metrics import r2_score,accuracy_score
```

```python
y_pred = model.predict(x_test)
y_pred
```
